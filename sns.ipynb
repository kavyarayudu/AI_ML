```
import seaborn as sns
sns.set(style="dark")
tips=sns.load_dataset("tips")
print(tips.columns)
print(tips.head)
sns.boxplot(x=tips["size"])
```

```
import seaborn as sns
sns.set(style="ticks")
tips=sns.load_dataset("tips")
sns.stripplot(x=tips["total_bill"])
```

```
ax=sns.boxplot(x="day",y="total_bill",data=tips)
```

```
sns.boxplot(x="day", y="total_bill", hue="smoker",
                 data=tips, palette="husl")
```

```
sns.boxplot(x="day", y="total_bill", hue="smoker",
                 data=tips, palette="GnBu",linewidth=1.5)
```

```
import seaborn as sns
iris=sns.load_dataset("iris")
ax = sns.boxplot(data=iris, orient="h", palette="Set3")

```

```
sns.boxplot(x="time",y="tip",data=tips,
            order=["Dinner","Lunch"])
```

```
sns.boxplot(x="day",y="total_bill",data=tips,color="blue")
sns.swarmplot(x="day",y="total_bill",data=tips,color="pink")
```
